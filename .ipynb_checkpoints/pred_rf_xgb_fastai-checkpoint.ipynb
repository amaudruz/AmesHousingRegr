The evaluation metric is the rmse of the predicted output vs real output, so we can already simply work with the transform the dependent variable with the logarithm.

```python
df['SalePrice'] = np.log1p(df['SalePrice'])
```

```python
df.head()
```

```python
X, y, na_dict = proc_df(df.copy(), y_fld='SalePrice')
```

```python
train_X, test_X, train_y, test_y = train_test_split(X.as_matrix(), y, test_size=0.25)
```

```python
def my_rmse(y, y_pred) :
    return np.sqrt(((y- y_pred)**2).mean())
```

```python
def score(model) :
    return model.score(train_X, train_y), model.score(test_X, test_y), my_rmse(model.predict(train_X), train_y), my_rmse(model.predict(test_X), test_y)
```

## Random forest

```python
rd = RandomForestRegressor(n_jobs=-1)
```

```python
rd.fit(train_X, train_y)
```

```python
score(rd)
```

```python
rd_s = RandomForestRegressor(n_jobs=-1, n_estimators=1000)
```

```python
rd_s.fit(train_X, train_y)
```

```python
score(rd_s)
```

## Xgboost

```python
from xgboost import XGBRegressor
```

```python
xg = XGBRegressor()
```

```python
xg.fit(train_X, train_y, verbose=False)
```

```python
score(xg)
```

```python
xg_s = XGBRegressor(n_estimators=100, learning_rate=0.1, subsample=0.8,
           colsample_bytree=0.8, max_depth=5, min_child_weight=2)
```

```python
xg_s.fit(train_X, train_y, early_stopping_rounds=5, eval_set=[(test_X, test_y)] )
```

```python
score(xg_s)
```
